```python
#dropping redundencies

student = df.drop(['race/ethnicity', 'parental level of education'], axis=1)
student.head()
```

```python
# relationship analysis

correlation = student.corr(numeric_only=True)
```

```python
sns.heatmap(correlation, xticklabels=correlation.columns, yticklabels=correlation.columns, annot=True)
```

```python
sns.pairplot(student)
```

```python
sns.relplot(x='math score', y='reading score', hue='gender', data=student)
```

```python
sns.relplot(x='math score', y='reading score', hue='lunch', data=student)
```

```python
sns.displot(student['math score'], kde=True, linewidth=0)
```

```python
sns.displot(student['reading score'], kde=True, linewidth=0)
```

```python
sns.displot(student['writing score'], kde=True, linewidth=0, bins=10)
```
